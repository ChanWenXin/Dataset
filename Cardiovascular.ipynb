```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['age_year'], bins=20, kde=True)  # Increase bins for better distribution
plt.title("Age Distribution of Patients")
plt.xlabel("Age (Years)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['age_year'])
plt.title("Age Distribution of Patients (Boxplot)")
plt.xlabel("Age (Years)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a density plot for Age vs. Cardiovascular Disease
plt.figure(figsize=(8, 5))
sns.kdeplot(df[df['cardio'] == 0]['age_year'], label="No Disease", shade=True, color="blue")
sns.kdeplot(df[df['cardio'] == 1]['age_year'], label="Has Disease", shade=True, color="red")

# Add titles and labels
plt.title("Age vs. Cardiovascular Disease (Density Plot)")
plt.xlabel("Age (Years)")
plt.ylabel("Density")
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a boxplot for systolic blood pressure (ap_hi)
sns.boxplot(x=df["cardio"], y=df["ap_hi"], palette=["blue", "red"])

# Add titles and labels
plt.title("Systolic Blood Pressure vs. Cardiovascular Disease")
plt.xlabel("Cardiovascular Disease (0 = No, 1 = Yes)")
plt.ylabel("Systolic Blood Pressure (ap_hi)")

# Show the plot
plt.show()

# Now plotting for diastolic blood pressure (ap_lo)
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["cardio"], y=df["ap_lo"], palette=["blue", "red"])

# Add titles and labels
plt.title("Diastolic Blood Pressure vs. Cardiovascular Disease")
plt.xlabel("Cardiovascular Disease (0 = No, 1 = Yes)")
plt.ylabel("Diastolic Blood Pressure (ap_lo)")

# Show the plot
plt.show()
```
